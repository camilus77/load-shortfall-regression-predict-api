If we could recall, one feature was not in numerical format. Let us have peek at it.

```python
final_data['Valencia_wind_deg'].head()
```

It seems that the problem is with the 'Level_' appended to the feature. let us extract the non-alphabets and convert

```python
final_data['Valencia_wind_deg'] = final_data['Valencia_wind_deg'].str.extract('(\d+)')
final_data['Valencia_wind_deg'] = pd.to_numeric(final_data['Valencia_wind_deg'])
```

...Seems as it looks good. We will perform similar task on the Seville_pressure feature

```python
final_data['Seville_pressure']=final_data['Seville_pressure'].str.extract('(\d+)')
final_data['Seville_pressure']=pd.to_numeric(final_data['Seville_pressure'])
```

....Great!!

We also observed that the time column is an object. Let us change it to date-time.
We will also love create new features (Month, Hour) from the time column.

We also wish to break down months into 4 seasons i.e `Winter(June-Aug)` as 1, and `Spring(Sep-Nov)` as 2, 
`Summer(Dec-mar)` as 3,and `Autumn(Apr-May)` as 4

```python
def Season(monthID):
    if monthID>=6 and monthID<=8:
        return 1
    elif monthID>=9 and monthID<=11:
        return 2
    elif monthID==12 or monthID<=2:
        return 3
    else:
        return 4

```

We wish to break down the day into five i.e `Dusk (1-5)` as 1, `Morning (6-12)` as 2, `Afternoon (13-15)` as 3, `Evening (16-19)` as 4, `Night(20-23.59)` as 5

```python

def Time_of_day(HourID):
    if HourID<=5:
        return 1
    elif HourID>=6 and HourID<=12:
        return 2
    elif HourID>=13 and HourID<=15:
        return 3
    elif HourID>=16 and HourID<=19:
        return 4
    else:
        return 5
```

```python
#We now change the datetime
final_data['time']=pd.to_datetime(final_data['time'])

# Create new columns - year/ month/ day / week_day/hour
final_data['year'] = final_data['time'].dt.year
final_data['month'] = final_data['time'].dt.month
final_data['day'] = final_data['time'].dt.day
final_data['hour'] = final_data['time'].dt.hour
final_data['week_day'] = final_data['time'].dt.dayofweek
#final_data[['year','month', 'day', 'week_day', 'hour']]

final_data['Season']=final_data['month'].apply(lambda x:Season(x))
final_data['Time_of_day']=final_data['hour'].apply(lambda x:Time_of_day(x))
final_data=final_data.drop(['time','month','hour'], axis=1)

```

We now wish to design a collapse the wind_deg features for all cities into a single feature

```python
# Create two new columns - Pressure and Wind_degree 
final_data['Wind_deg'] = final_data[['Valencia_wind_deg', 'Bilbao_wind_deg', 'Barcelona_wind_deg']].mean(axis=1)
final_data['Pressure'] = final_data[['Madrid_pressure', 'Valencia_pressure', 'Bilbao_pressure', 'Barcelona_pressure']].mean(axis=1)

#  Drop individual pressure and wind degree columns
final_data = final_data.drop(['Valencia_wind_deg', 'Bilbao_wind_deg', 'Barcelona_wind_deg', 'Madrid_pressure', 
              'Valencia_pressure', 'Bilbao_pressure', 'Seville_pressure', 'Barcelona_pressure'], axis =1)
```

```python
final_data.head()
```

It seems as if the quite features are many. We wish to remove the redundant features. 
In this case, we will check those with decent correlation with the 'load_shortfall_3h'
as well as those with significance (using p-value)

```python
final_corr=final_data[:len(train)]
# Calculate correlations between predictor variables and the response variable
corrs = final_corr.corr()['load_shortfall_3h'].sort_values(ascending=False)


# Build a dictionary of correlation coefficients and p-values
dict_cp = {}

column_titles = [col for col in corrs.index if col!= 'load_shortfall_3h']
for col in column_titles:
    p_val = round(pearsonr(final_corr[col], final_corr['load_shortfall_3h'])[1],6)
    dict_cp[col] = {'Correlation_Coefficient':corrs[col],
                    'P_Value':p_val}

df_cp = pd.DataFrame(dict_cp).T
df_cp_sorted = df_cp.sort_values('P_Value')
df_cp_sorted[df_cp_sorted['P_Value']<0.1]
```

It seems all the features are significant. They all have p-values below 0.05 i.e 5% significance. Now let us check for 
collinearity.

```python
# We are looking for rows and columns where correlation coefficients > 0.9 or <-0.9
corr = final_corr.drop('load_shortfall_3h', axis = 1).corr()
# Find rows and columnd where correlation coefficients > 0.9 or <-0.9
corr[np.abs(corr) > 0.9]
```

From the cross-correlation, it is observed that temp,temp_min and temp_max for the all the cities are highly correlated.
This could lead to multi-collinearity.

The significance check earlier done shows that apart from madrid, temp_min is the most significant feature.
For this reason, we drop temp and temp_max for all the cities

```python
final_data=final_data.drop(['Barcelona_temp', 'Barcelona_temp_max','Madrid_temp', 'Madrid_temp_max','Seville_temp', 'Seville_temp_max','Bilbao_temp', 'Bilbao_temp_max','Valencia_temp', 'Valencia_temp_max'], axis=1)
```

The random forest model is  built on idea of decision trees. In this case, our model contains many decision trees representing a distinct instance of the umbundling the data input into a ‘random forest’. The random forest technique then the instances individually, selecting the one with the majority of votes as the selected prediction.

### Merits
- The random forest technique emphasizes feature selection — weighing certain features as more important than others. This is very unlike other regression models who assume a linear or semi-linear relationship between variables.

- The method also adopts bootstrapping which when combined with the unpruned nodes allows optimal nodes to be sampled from the total nodes in the tree to form the optimal splitting feature.
- Forming optimal features lowers the correlation and hence, the variance of the regression trees. It improves the predictive capability of distinct trees in the forest. The sampling procedure also increases independence among individual trees.
